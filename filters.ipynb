```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
from tensorflow.keras.datasets import mnist
```

### Apply a filter to simple image

```python
# read in image (make sure you downloaded the images supplied with this notebook)
square = mpimg.imread('black_square.jpg')

# ensure the image is grayscale
gray_square = cv2.cvtColor(square, cv2.COLOR_RGB2GRAY)

# filter for vertical edge detection
horizontal = np.array([[-1, -2, -1], 
                       [ 0, 0, 0], 
                       [ 1, 2, 1]])

# filter for horizontal edge detection
vertical = np.array([[ 1, 0, -1], 
                     [ 2, 0, -2], 
                     [ 1, 0, -1]])

# filter for horizontal edge detection
'''diag = np.array([[ 0, -2, -1], 
                 [ 2, 0, -1], 
                 [ 1, 1, 0]])'''

# apply the filters to the image using filter2D from opencv package  
filtered_h = cv2.filter2D(gray_square, -1, horizontal)
filtered_v = cv2.filter2D(gray_square, -1, vertical)
#filtered_d = cv2.filter2D(gray_square, -1, diag)

# plot the results
f, axarr = plt.subplots(1, 3, sharey=True, figsize=(12, 12))
axarr[0].set_title('Original')
axarr[0].imshow(gray_square, cmap='gray')
axarr[1].set_title('"Horizontal Filter"')
axarr[1].imshow(filtered_h, cmap='gray')
axarr[2].set_title("Vertical Filter")
axarr[2].imshow(filtered_v, cmap='gray')
'''axarr[3].set_title("Diag Filter")
axarr[3].imshow(filtered_d, cmap='gray')'''
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

one = train_images[10]

# Apply edge detection filters to the grayscale image
horizontal = np.array([[-1, -2, -1], 
                       [0, 0, 0], 
                       [1, 2, 1]])

vertical = np.array([[1, 0, -1], 
                     [2, 0, -2], 
                     [1, 0, -1]])

diag = np.array([[0, -2, -1], 
                 [2, 0, -1], 
                 [1, 1, 0]])

# Apply the filters to the image using filter2D from the OpenCV package
filtered_h = cv2.filter2D(one, -1, horizontal)
filtered_v = cv2.filter2D(one, -1, vertical)
filtered_d = cv2.filter2D(one, -1, diag)

# Plot the results
f, axarr = plt.subplots(1, 4, sharey=True, figsize=(12, 12))
axarr[0].set_title('Original')
axarr[0].imshow(one, cmap='gray')
axarr[1].set_title('Horizontal Filter')
axarr[1].imshow(filtered_h, cmap='gray')
axarr[2].set_title('Vertical Filter')
axarr[2].imshow(filtered_v, cmap='gray')
axarr[3].set_title('Diagonal Filter')
axarr[3].imshow(filtered_d, cmap='gray')
```

### Apply filters to a regular image

```python
Denmark = mpimg.imread('Denmark.jpg')
Denmark_gray = cv2.cvtColor(Denmark, cv2.COLOR_RGB2GRAY)

f, axarr = plt.subplots(1, 2, sharey=True, figsize=(12, 12))
axarr[0].set_title('Original')
axarr[0].imshow(Denmark, cmap='gray')
axarr[1].set_title('Grayscale')
axarr[1].imshow(Denmark_gray, cmap='gray')
```

```python
# filter for vertical edge detection
horizontal = np.array([[-1, -2, 1], 
                       [ 0, 0, 0], 
                       [ 1, 2, 1]])

# filter for horizonatl edge detection
vertical = np.array([[ -1, 0, 1], 
                     [ -2, 0, 2], 
                     [ -1, 0, 1]])

diag = np.array([[ 0, -2, -1], 
                 [ 2, 0, -1], 
                 [ 1, 1, 0]])

filtered_h = cv2.filter2D(Denmark_gray, -1, horizontal)
filtered_v = cv2.filter2D(Denmark_gray, -1, vertical)
filtered_d = cv2.filter2D(Denmark_gray, -1, diag)

f, axarr = plt.subplots(1, 4, sharey=True, figsize=(12, 12))
axarr[0].set_title('Original')
axarr[0].imshow(Denmark_gray, cmap='gray') 
axarr[1].set_title('Horizontal Filter')
axarr[1].imshow(filtered_h, cmap='gray')
axarr[2].set_title('Vertical Filter')
axarr[2].imshow(filtered_v, cmap='gray')
axarr[3].set_title("Diag Filter")
axarr[3].imshow(filtered_d, cmap='gray')
```
